## Description of problem  12: Highly divisible triangular number **[link](https://projecteuler.net/problem=12)**
    
The sequence of triangle numbers is generated by adding the natural numbers. 
So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28.

The first ten terms would be:
1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

| Triangle Number | Factors |
|:---:|:---|
| **1** | 1 |
| **3** | 1,3 |
| **6** | 1,2,3,6 |
| **10** | 1,2,5,10 |
| **15** | 1,3,5,15 |
| **21** | 1,3,7,21 |
| **28** | 1,2,4,7,14,28 |

<span style="display:none"></span>

We can see that 28 is the first triangle number to have over five divisors.
What is the value of the first triangle number to have over five hundred divisors?

## Brainstorm

The challenge over here is consisting of two challenges:
1. generating triangle numbers
2. generating the factors

*So on first sight it simple, but now let's see what we can do based on mathematics*

* Regarding the first challenge:
Most of us known that summing up all numbers to a certain number (we call it *n* from now on) is easly done by mutiplying the 
number and number plus one (*n+1*) and dividing it by two, this gives you the following formula's to work with:  
 1 + 2 + 3 + ... + (n-2) + (n-1) + n = $\sum_{i=1}^n n$ = n*(n+1)/2  
This enables us to calculate the triangle number instead of adding up all the numbers in a loop to generated the total. 


* Regarding the second challenge:
Generating the factors can be the tricky part however mathematics is again your interpretation friend on this one. Instead of factoring all the factors until the square root of the number we are calculating. We could only generate the prime factors and their occurrence (e.a. power of the prime factor).
By removing the 1 and the number you are calculating, you can use the multiply of the occurrence +1 of the first prime factor with the occurrence +1 of the second prime factor and so on. Let's use the 7th triangle number having value 28 as an example:

| Triangle Number | Factors | Prime Factors | Number of Factors |
|:---:|:---|:---|:---|
| **28** | 1,2,4,7,14,28 | 1,$2^2$,$7^1$,28 | (2+1).(1+1) = 6 |
